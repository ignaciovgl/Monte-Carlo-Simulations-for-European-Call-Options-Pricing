```python
import numpy as np
import matplotlib.pyplot as plt

S_0 = 100  # Initial stock price
K = 110    # Strike price
T = 1.0    # Time to maturity (1 year)
r = 0.05   # Risk-free rate (5%)
sigma = 0.2  # Volatility (20%)
N = 10000   # Number of simulations
M = 365     # Number of time steps (daily steps)

dt = T / M  # Time step

# Initialize price paths: rows are time steps, columns are paths
S = np.zeros((M + 1, N))
S[0] = S_0

# Generate paths
for t in range(1, M + 1):
    Z = np.random.normal(0, 1, N)  # Standard normal random numbers
    S[t] = S[t - 1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
    
# Final prices at maturity
S_T = S[-1]
# Calculate payoffs
payoffs = np.maximum(S_T - K, 0)

print(payoffs)
option_price = np.exp(-r * T) * np.mean(payoffs)
print(f"Simulated European Call Option Price: {option_price:.2f}")
```

```python
import numpy as np

S_0 = 11789  # Initial stock price
K = 12000    # Strike price
T = 1.0    # Time to maturity (1 year)
r = 0.0041   # Risk-free rate (5%)
sigma = 0.122  # Volatility (20%)
N = 10000   # Number of simulations
M = 365   # Number of time steps (daily steps)

dt = T / M  # Time step

# Initialize price paths: rows are time steps, columns are paths
S = np.zeros((M + 1, N))
S[0] = S_0

# Generate paths
for t in range(1, M + 1):
    Z = np.random.normal(0, 1, N)  # Standard normal random numbers
    S[t] = S[t - 1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
    
# Final prices at maturity
S_T = S[-1]

# Calculate payoffs
payoffs = np.maximum(S[-1] - K, 0)

option_price = np.exp(-r * T) * np.mean(payoffs)
print(f"Simulated European Call Option Price: {option_price:.2f}")

# Plot a subset of paths
plt.figure(figsize=(10, 6))
for i in range(20):  # Plot only 20 paths for clarity
    plt.plot(S[:, i], lw=1, alpha=0.7)

# Add labels and title
plt.title("Simulated Asset Price Paths (Monte Carlo)", fontsize=14)
plt.xlabel("Time Steps (Days)", fontsize=12)
plt.ylabel("Asset Price", fontsize=12)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(S_T, bins=50, alpha=0.75, color='blue', edgecolor='black')
plt.title("Distribution of Asset Prices at Maturity (S_T)", fontsize=14)
plt.xlabel("Asset Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()
```
